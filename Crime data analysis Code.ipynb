```python
import pandas as pd
import numpy as np
import matplotlib as mpl # optional (here)
import matplotlib.pyplot as plt
from ipywidgets import *
import seaborn as sns # Optional, will only affect the color of bars and the grid

from ipywidgets import widgets, interactive, interact
%matplotlib inline  
import pandas as pd
x=pd.read_csv("C:\Project 1\Crimes_-_2001_to_present1.csv",sep=",")
def plotit(CrimeType,Month):
    if Month!=0 and CrimeType!="All":
        fig,(ax1,ax2)=plt.subplots(nrows=1, ncols=2)
        targettime=pd.DataFrame(x.groupby(["Primary Type","Month"]).size())
        targettime.reset_index(level=["Primary Type","Month"], inplace=True)
        targettime.columns=["Primary Type","Month","CrimeCount"]
        targettime1=targettime[targettime["Primary Type"]==CrimeType]

        targettime1.plot(x=["Month"],y=["CrimeCount"],kind="barh",ax=ax1,rot=90,alpha=0.5,fontsize=10,figsize=(12,4),color="red")
        ax1.set_title("Type of Crime in month: "+str(Month),fontsize=15)
        ax1.set_ylabel("Months",fontsize=15)
        ax1.set_xlabel("Crime Counts",fontsize=15)
        #ax1.gcf().subplots_adjust(bottom=0.15)
        
       
        
        targettime=pd.DataFrame(x.groupby(["Primary Type","Month","Hourd"]).size())
        targettime.reset_index(level=["Primary Type","Month","Hourd"], inplace=True)
        targettime.columns=["Primary Type","Month","Hourd","CrimeCount"]
        targettime1=targettime[(targettime["Month"]==Month) & (targettime["Primary Type"]==CrimeType)]
        targettime1.plot(x=["Hourd"],y=["CrimeCount"],kind="bar",ax=ax2,alpha=0.5,fontsize=10,color="blue",figsize=(12,4))
        ax2.set_title("Time wise crime in Month: "+str(Month),fontsize=15)
        
        ax2.set_xlabel("Time",fontsize=15)
        ax2.set_ylabel("Crime Counts",fontsize=15)
        #ax2.gcf().subplots_adjust(bottom=0.15)
        plt.subplots_adjust(wspace=0.7)
        plt.show()
CrimeType = widgets.Dropdown(
    options=['All'] + list(x['Primary Type'].unique()),
    value='All',
    description='Crime Type:',
)

interact(plotit,CrimeType=CrimeType,Month=(1,10,1))
#interactive(plotit,CrimeType=CrimeType,Month=Month)    
```

```python
#Zipcode and Time
import pandas as pd
import numpy as np
import matplotlib as mpl # optional (here)
import matplotlib.pyplot as plt
import seaborn as sns # Optional, will only affect the color of bars and the grid

from ipywidgets import widgets, interactive
%matplotlib inline  
import pandas as pd
x=pd.read_csv("C:\Project 1\With_Zip.csv",sep=",")
def plotit(CrimeType,ZipCodes):
    if CrimeType!="All":
        fig,(ax1,ax2)=plt.subplots(nrows=1, ncols=2)
        targettime=pd.DataFrame(x.groupby(["Primary Type","ZipCodes"]).size())

        targettime.reset_index(level=["Primary Type","ZipCodes"], inplace=True)
        targettime.columns=["Primary Type","ZipCodes","CrimeCount"]
        targettime1=targettime[targettime["Primary Type"]==CrimeType]

        targettime1=targettime1.sort("CrimeCount",ascending=False)
        targettime1.head(5).plot(x=["ZipCodes"],y=["CrimeCount"],ax=ax1,kind="bar",alpha=0.5,fontsize=12,figsize=(12,4),color="blue")
        ax1.set_title("Top 5 Zipcode Where " + CrimeType + " is High",fontsize=10)
        ax1.set_xlabel("ZipCodes",fontsize=12)
        ax1.set_ylabel("Crime Counts",fontsize=12)
        #plt.gcf().subplots_adjust(bottom=0.15)

        if (ZipCodes)>60000:
            targettime=pd.DataFrame(x.groupby(["Primary Type","ZipCodes","Hour"]).size())
            targettime.reset_index(level=["Primary Type","ZipCodes","Hour"], inplace=True)
            targettime.columns=["Primary Type","ZipCodes","Hour","CrimeCount"]
            targettime1=targettime[(targettime["ZipCodes"]==ZipCodes) & (targettime["Primary Type"]==CrimeType)]
            targettime1.plot(x=["Hour"],y=["CrimeCount"],kind="area",ax=ax2,alpha=0.5,fontsize=12,color="orange",figsize=(12,4))
            ax2.set_title("Crime Statistics based on Time of ZipCode: "+str(ZipCodes),fontsize=10)
            ax2.set_xlabel("Time",fontsize=12)
            ax2.set_ylabel("Crime Counts",fontsize=12)
            #ax2.gcf().subplots_adjust(bottom=0.15)
            plt.subplots_adjust(wspace=0.7)
            plt.show()
CrimeType = widgets.Dropdown(
    options=['All'] + list(x['Primary Type'].unique()),
    value='All',
    description='Crime Type:',
)
ZipCodes = widgets.BoundedFloatText(
    value=0,
    min=0,
    max=x.ZipCodes.max(),
    #options=list(targettime['Month'].unique()),
    description='Zip Codes:',
)
interactive(plotit,CrimeType=CrimeType,ZipCodes=ZipCodes)    
```

```python
#Zipcodes and Crime Type
import pandas as pd
import numpy as np
import matplotlib as mpl # optional (here)
import matplotlib.pyplot as plt
import seaborn as sns # Optional, will only affect the color of bars and the grid

from ipywidgets import widgets, interactive
%matplotlib inline  
import pandas as pd
x=pd.read_csv("C:\Project 1\With_Zip.csv",sep=",")
def plotit(n,ZipCodes):
    if n!=0:
        fig,(ax1,ax2)=plt.subplots(nrows=1, ncols=2)
        targettime=pd.DataFrame(x.groupby(["ZipCodes"]).size())

        targettime.reset_index(level=["ZipCodes"], inplace=True)
        targettime.columns=["ZipCodes","CrimeCount"]
        targettime=targettime.sort("CrimeCount",ascending=False)
        y=int(n)
        targettime.head(y).plot(x=["ZipCodes"],y=["CrimeCount"],kind="bar",ax=ax1,alpha=0.5,fontsize=12,figsize=(12,4),color="red")
        ax1.set_title("Top "+str(y)+" Zipcodes where Crime Rate is Highest",fontsize=12)
        ax1.set_xlabel("ZipCodes",fontsize=12)
        ax1.set_ylabel("Crime Counts",fontsize=12)
        #plt.gcf().subplots_adjust(bottom=0.15)
        if (ZipCodes)>60000:
            targettime=pd.DataFrame(x.groupby(["Primary Type","ZipCodes"]).size())
            targettime.reset_index(level=["Primary Type","ZipCodes"], inplace=True)
            targettime.columns=["Primary Type","ZipCodes","CrimeCount"]
            targettime1=targettime[(targettime["ZipCodes"]==ZipCodes)]
            targettime1=targettime1.sort("CrimeCount",ascending=False)
            targettime1.head(5).plot(x=["Primary Type"],y=["CrimeCount"],ax=ax2,kind="line",rot=90,alpha=0.5,fontsize=12,color="green",figsize=(12,4))
            ax2.set_title("Crimes that occur more at " + str(ZipCodes),fontsize=12)
            ax2.set_xlabel("Time",fontsize=12)
            ax2.set_ylabel("Crime Counts",fontsize=12)
            plt.subplots_adjust(wspace=0.7)
            #plt.gcf().subplots_adjust(bottom=0.15)
            plt.show()
n = widgets.BoundedFloatText(
    value=0,
    min=0,
    max=50,
    #options=list(targettime['Month'].unique()),
    description='N:',
)
ZipCodes = widgets.BoundedFloatText(
    value=0,
    min=0,
    max=x.ZipCodes.max(),
    #options=list(targettime['Month'].unique()),
    description='Zip Codes:',
)
interactive(plotit,n=n,ZipCodes=ZipCodes)    
```

```python
import pandas as pd
import numpy as np
import matplotlib as mpl # optional (here)
import matplotlib.pyplot as plt
import seaborn as sns # Optional, will only affect the color of bars and the grid

from ipywidgets import widgets, interactive
%matplotlib inline  
import pandas as pd
#x=pd.read_csv("C:\Project 1\Crimes_-_2001_to_present11.csv",sep=",")
def plotit(Year):
    if Year!='All':
        fig,(ax1,ax2,ax3)=plt.subplots(nrows=1, ncols=3)
        targettime=pd.DataFrame(x.groupby(["Primary Type","Year","DayofWeek"]).size())
        targettime.reset_index(level=["Primary Type","Year","DayofWeek"], inplace=True)
        targettime.columns=["Primary Type","Year","DayofWeek","CrimeCount"]
        targettime1=targettime[targettime["Year"]==Year]
        targettime1=targettime1.sort("CrimeCount",ascending=False)
        targettime1.index=np.arange(0,len(targettime1))
        targettime1.head(3).plot(x=["Primary Type"],y=["CrimeCount"],ax=ax1,kind="bar",alpha=0.5,fontsize=12,color="red",figsize=(12,4))
        ax1.set_title("Top 3 Crimes in a Year "+str(Year),fontsize=12)
        ax1.set_xlabel("Crime Type",fontsize=12)
        ax1.set_ylabel("Crime Counts",fontsize=12)
        #plt.gcf().subplots_adjust(bottom=0.15)
        targettime=pd.DataFrame(x.groupby(["Year","Arrest"]).size())
        targettime.reset_index(level=["Year","Arrest"], inplace=True)
        targettime.columns=["Year","Arrest","CrimeCount"]
        targettime1=targettime[targettime["Year"]==Year]
        targettime1.head(3).plot(x=["Arrest"],y=["CrimeCount"],ax=ax2,kind="bar",alpha=0.5,fontsize=12,color="blue",figsize=(12,4))
        ax2.set_title("Arested vs Not Arrested in "+str(Year),fontsize=12)
        ax2.set_xlabel("Arrested?",fontsize=12)
        ax2.set_ylabel("Counts",fontsize=12)
        
        targettime1=pd.DataFrame(x.groupby(["Year","DayofWeek"]).size())
        targettime1.reset_index(level=["Year","DayofWeek"], inplace=True)
        targettime1.columns=["Year","DayofWeek","CrimeCount"]
        targettime1=targettime1.sort("CrimeCount",ascending=False)
        targettime1.index=np.arange(0,len(targettime1))
        targettime1=targettime1[targettime1.Year==Year]
        targettime1.plot(x=["DayofWeek"],y=["CrimeCount"],ax=ax3,kind="bar",alpha=0.5,fontsize=10,color="green",figsize=(12,4))
        ax3.set_title("Day of week in "+str(Year),fontsize=12)
        ax3.set_xlabel("Crime Type",fontsize=12)
        ax3.set_ylabel("Crime Counts",fontsize=12)
        plt.subplots_adjust(wspace=0.7)
        #plt.gcf().subplots_adjust(bottom=0.15)
        plt.show()
        
Year = widgets.BoundedFloatText(
    #options=['All'] + list(x['Year'].unique()),
    value=x.Year.max(),
    min=0,
    max=x.Year.max(),
    description='Year:',
)
interactive(plotit,Year=Year)    
```

```python
import pandas as pd
import numpy as np
import matplotlib as mpl # optional (here)
import matplotlib.pyplot as plt
import seaborn as sns # Optional, will only affect the color of bars and the grid

from ipywidgets import widgets, interactive
%matplotlib inline  
import pandas as pd
x=pd.read_csv("C:\Project 1\With_Zip.csv",sep=",")
x["Location Description"][x["Location Description"]==0]="NON-RESIDENCE"
x["Location Description"][x["Location Description"]==1]="RESIDENCE"
def plotit(CrimeType,Time):
    if CrimeType!="All":
        fig,(ax1,ax2)=plt.subplots(nrows=1, ncols=2)
        targettime=pd.DataFrame(x.groupby(["Primary Type","Hour"]).size())
        targettime.reset_index(level=["Primary Type","Hour"], inplace=True)
        targettime.columns=["Primary Type","Hour","CrimeCount"]
        targettime1=targettime[targettime["Primary Type"]==CrimeType]
        targettime1.plot(x=["Hour"],y=["CrimeCount"],kind="bar",ax=ax1,alpha=0.5,fontsize=10,figsize=(10,4))
        ax1.set_title("Crime rate for "+CrimeType+" based on time",fontsize=10)
        ax1.set_xlabel("Time",fontsize=12)
        ax1.set_ylabel("Crime Counts",fontsize=12)
        #plt.gcf().subplots_adjust(bottom=0.15)

        if Time!="All":
            targettime=pd.DataFrame(x.groupby(["Primary Type","Location Description","Hour"]).size())
            targettime.reset_index(level=["Primary Type","Location Description","Hour"], inplace=True)
            targettime.columns=["Primary Type","Location Description","Hour","CrimeCount"]
            targettime1=targettime[(targettime["Hour"]==Time) & (targettime["Primary Type"]==CrimeType)]
            targettime1.plot(x=["Location Description"],y=["CrimeCount"],kind="bar",ax=ax2,alpha=0.5,fontsize=10,color="yellow",figsize=(10,4))
            ax2.set_title("Crime Statistics based on Location in " + Time,fontsize=10)
            ax2.set_xlabel("Location",fontsize=12)
            ax2.set_ylabel("Crime Counts",fontsize=12)
            plt.subplots_adjust(wspace=0.7)
            #plt.gcf().subplots_adjust(bottom=0.15)
            plt.show()
CrimeType = widgets.Dropdown(
    options=['All'] + list(x['Primary Type'].unique()),
    value='All',
    description='Crime Type:',
)
Time = widgets.Dropdown(
    options=['All'] + list(x['Hour'].unique()),
    value='All',
    description='Time:',
)
interactive(plotit,CrimeType=CrimeType,Time=Time)    
```
